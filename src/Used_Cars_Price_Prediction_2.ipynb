```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

```python
df_no_price = df.drop(['price'], axis = 1)
```

```python
X = df_no_price.iloc[:,]
calc_vif(X)
```

```python
df_no_price_size = df_no_price.drop(['size'], axis = 1)
```

```python
X = df_no_price_size.iloc[:,]
calc_vif(X)
```

#chi2

```python
X = df_no_price_size
y = df['price']
```

```python
df_no_price_size.columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)
```

```python
X_train.isnull().sum().to_frame()
```

```python
print('Train', X_train.shape, y_train.shape)
print('Test', X_test.shape, y_test.shape)
```

```python
# configure to select all features
fs = SelectKBest(score_func=f_classif, k='all')
# learn relationship from training data
fs.fit(X_train, y_train)
# transform train input data
X_train_fs = fs.transform(X_train)
# transform test input data
X_test_fs = fs.transform(X_test)
```

```python
from matplotlib import pyplot
```

```python
# what are scores for the features
for i in range(len(fs.scores_)):
	print('Feature %d: %f' % (i, fs.scores_[i]))
# plot the scores
plt.figure(figsize=(14,10))
plt.bar([i for i in range(len(fs.scores_))], fs.scores_)
plt.show()
```

#Model

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
from sklearn.linear_model import LinearRegression
linear = LinearRegression()
linear.fit(X_train, y_train)
y_pred = linear.predict(X_test)

print(mean_squared_error(y_test, y_pred))
print((mean_squared_error(y_test, y_pred))**0.5)
round(r2_score(y_test, y_pred) * 100, 2)
```

```python
from sklearn.tree import DecisionTreeRegressor
decision_tree = DecisionTreeRegressor()
decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test)

print(mean_squared_error(y_test, y_pred))
print((mean_squared_error(y_test, y_pred))**0.5)
round(r2_score(y_test, y_pred) * 100, 2)
```

```python
from sklearn.ensemble import RandomForestRegressor
random_forest = RandomForestRegressor()
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)

print(mean_squared_error(y_test, y_pred))
print((mean_squared_error(y_test, y_pred))**0.5)
round(r2_score(y_test, y_pred) * 100, 2)
```

```python
from sklearn.ensemble import BaggingRegressor
bagging = BaggingRegressor()
bagging.fit(X_train,y_train)
y_pred = bagging.predict(X_test)

print(mean_squared_error(y_test, y_pred))
print((mean_squared_error(y_test, y_pred))**0.5)
round(r2_score(y_test, y_pred) * 100, 2)
```

```python
from xgboost import XGBRegressor
xgb_=XGBRegressor(random_state=42)
xgb_model=xgb_.fit(X_train,y_train)
y_pred=xgb_model.predict(X_test)
print(mean_squared_error(y_test, y_pred))
print((mean_squared_error(y_test, y_pred))**0.5)
round(r2_score(y_test, y_pred) * 100, 2)
```
